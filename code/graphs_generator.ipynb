```python

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.style.reload_library()
#plt.style.use(['classic'])
plt.style.use(['science',"ieee"])
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
s2 = np.cos(2 * np.pi * t)

plt.figure(figsize=(5.9,3.646))
plt.plot(t, s)
plt.plot(t, s2)
plt.xlabel("time (s)")
plt.ylabel("Voltage (mV)")
plt.title("Simple plot $\\frac{\\alpha}{2}$", pad=7)
plt.grid(False)
plt.savefig("../plot_data/test.eps")
```
